```python
import cohere
import os
co = cohere.ClientV2(os.getenv("COHERE_API_KEY"))
```

```python
prompt = "You are a helpful customer support assistant for CameraLK. " \
         "Help customers choose cameras, lenses, and accessories. " \
         "If the question is about product recommendations, suggest suitable cameras or accessories. " \
         "Do Not answer any questions other than the ones related to cameras, lenses, and accessories. " \
         "Keep the answers short" \
         
#chat history
messages = [{'role': 'system', 'content': prompt}
        ]         

print('Welcome to CaemeraLK customer support! How can I assist you today?')

while True:
    message = input('You: ')

    if message.lower() in ['exit', 'quit']:
        print('Thank you for contacting CameraLK customer support. Have a great day!')
        break

    #user msg to history
    messages.append({"role": "user", "content": message})

    response = co.chat_stream(
        model="command-a-03-2025",
        messages=messages,
    )

    bot_reply = ""
    print("Assistant: ", end="")

    for event in response:
        if event.type == "content-delta":
            text = event.delta.message.content.text
            bot_reply += text
            print(text, end="")

    print("\n")

    #bot reply to history
    messages.append({"role": "assistant", "content": bot_reply})

    #memory limit
    if len(messages) > 20:
        messages = [messages[0]] + messages[-19:]
```

```python
#chat history
for message in messages:
    print(message,"\n")
```
