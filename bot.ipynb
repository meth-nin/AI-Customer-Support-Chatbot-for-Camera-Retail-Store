```python
import cohere
import os
co = cohere.ClientV2(os.getenv("COHERE_API_KEY"))
```

```python
message = input("Enter your message: ")

prompt = "You are a helpful customer support assistant for CameraLK. " \
         "Help customers choose cameras, lenses, and accessories. " \
         "If the question is about product recommendations, suggest suitable cameras or accessories. " \
         "Do Not answer any questions other than the ones related to cameras, lenses, and accessories. " \
         "Keep the answers short" \

messages = [{'role': 'system', 'content': prompt},
            {'role': 'user', 'content': message}]

response = co.chat_stream(
    model= "command-a-03-2025",
    messages= messages
    )

for event in response:
    if event.type == 'content-delta':
        print(event.delta.message.content.text, end='', flush=True)
```
